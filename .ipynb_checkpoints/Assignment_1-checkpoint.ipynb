## Research Question

How does the prevalence of asthma and COPD vary across different states in the U.S. over time, and what factors (e.g., smoking rates, air quality) could be associated with these variations?

## Problem Statement

Asthma and chronic obstructive pulmonary disorder (COPD) are two significant chronic respiratory conditions affecting millions of Americans. These diseases are influenced by multiple factors such as environmental conditions, lifestyle choices, and access to healthcare. Understanding the geographic distribution and prevalence of these conditions across states is critical to improving public health initiatives. This analysis will explore the prevalence of asthma and COPD across different U.S. states and assess potential correlations with contributing factors like smoking rates and air quality.

## Data Description

U.S. Chronic Disease Indicators Dataset (https://catalog.data.gov/dataset/u-s-chronic-disease-indicators)

Last Updated: March 9, 2024

The dataset contains data on 115 indicators related to chronic diseases across U.S. states and territories. This dataset provides a uniform definition, collection, and reporting of chronic disease data important to public health practices in each area.
